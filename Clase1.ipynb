# <span style="color:red;">Gráfica de la ecuación $y=x^2$</span>

Vamos a Graficar una parábola de la forma: $y=x^2$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x
x = np.linspace(-10, 10, 400)  # Valores de -10 a 10
y = x**2  # Función y = x^2

# Crear la gráfica
plt.plot(x, y, label=r'$y = x^2$', color='b')

# Configuración de la gráfica
plt.axhline(0, color='black', linewidth=0.5)  # Eje X
plt.axvline(0, color='black', linewidth=0.5)  # Eje Y
plt.grid(True, linestyle='--', linewidth=0.5)  # Cuadrícula
plt.xlabel('x')8
plt.ylabel('y')
plt.title('Gráfica de $y = x^2$'2
plt.legend()

# Mostrar la gráfica
plt.show()
```

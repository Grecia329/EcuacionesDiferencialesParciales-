```python
import numpy as np

cartas = np.arange(52)
print(cartas)
```

```python
(cartas % 13) + 1
```

```python
Ej = 0

valores = (cartas % 13) + 1
p = (1/52)*(1/51)
for ci in valores:
  for cj in valores:
    if ci != cj:
      if ci+cj >= 15:
        Ej += p
      else:
        Ej -= p

print(Ej)
```

```python
import random

n = 10**6

suma = 0

for i in range(n):
  c1 = int(random.random()*52) + 1  # carta uno en [1,52]
  v1 = (c1 % 13) + 1                # valor en [1,13]
  mv = int(random.random()*51) + 1  # maximo valor en [1,51] que se sumarÃ¡ a c1
  c2 = ( (c1 +  mv) % 52 ) + 1      # carta dos entre [1,52] sin repetir c1
  v2 = (c2 % 13) + 1                # valor en [1,13]
  if v1 + v2 >= 15:
    suma += 1
  else:
    suma -= 1

print("La probabilidad estimada es: ", suma/n)
```

# ONDAS VIAJERAS


Grafique la función:
$$
f(x')=exp(-\frac{(x'-0.5)^2}{0.01})
$$ y suponga que esta representa una perturbación en una cuerda.


 <span style="color:magenta;"> a) Mostrar que al hacer el cambio de variable $x'=x-ct$ la pertubación se desplaza a la derecha.</span>

```python
import numpy as np
import matplotlib.pyplot as plt
```

Parámetros

```python
x = np.linspace(-1, 4, 500)
c = .6 # velocidad
t = 0.8 # tiempo
x0 = 1 # centro del pulso
sigma = 0.5
```

Función Gaussiana

```python
def f(xp):
    return np.exp(-((xp - x0)**2) / sigma)

f_derecha = f(x - c * t)
```

 Graficar

```python
plt.figure(figsize=(10, 5))

plt.plot(x, f_derecha, label='(a) Onda a la derecha', linestyle='--', color='red')

plt.title("Ondas viajeras")
plt.xlabel("x")
plt.ylabel("f(x, t)")
plt.legend()
plt.grid(True)
plt.show()
```

<span style="color:magenta;"> B) Mostrar que al hacer el cambio de variable \( x' = x + ct \), la perturbación se desplaza a la **izquierda**.</span>

```python
def f(xp):
    return np.exp(-((xp - x0)**2) / sigma)
```

```python
f_izquierda = f(x + c * t)
```

```python
plt.figure(figsize=(10, 6))


plt.plot(x, f_izquierda, label='(b) Onda a la izquierda', linestyle='-.')

plt.title("Ondas viajeras")
plt.xlabel("x")
plt.ylabel("f(x, t)")
plt.legend()
plt.grid(True)
plt.show()
```

<span style="color:magenta;">(c) Graficar la **superposición** de las ondas viajeras de los incisos **a)** y **b)**.</span>


```python
def f(xp):
    return np.exp(-((xp - x0)**2) / sigma)

# Incisos a), b), y c)
f_derecha = f(x - c * t)
f_izquierda = f(x + c * t)
f_total = f_derecha + f_izquierda
```

```python

plt.figure(figsize=(10, 5))
plt.plot(x, f_derecha, label='Onda a la derecha (x - ct)', color='blue', linestyle='--')
plt.plot(x, f_izquierda, label='Onda a la izquierda (x + ct)', color='green', linestyle='-.')
plt.plot(x, f_total, label='Superposición (a + b)', color='red', linewidth=2)

plt.title("Superposición de Ondas Gaussianas")
plt.xlabel("x")
plt.ylabel("f(x, t)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

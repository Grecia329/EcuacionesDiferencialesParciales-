# Práctica 1 (J) Juegos con Cartas

# Juego 1

```python
import numpy as np
import random
```

```python
cartas = np.arange(52)
print(cartas)
```

Obtener un número entero entre 0 y 51, y guardarlo dentro de una variable carta.

```python
carta = int(np.random.randint(0,52))

print(carta)
```

**Simular 1000 partidas**

```python
n = 1000000

juego = 0


for i in range(n):
  carta = int(np.random.randint(0,52))
  if carta < 26:
    juego += 1
  else:
    juego -= 1

# print(juego)
```

**Calcular la ganancia promedio:**

$$
\text{promedio} = \frac{\text{ganancia total}}{\text{# de juegos}}
$$

```python
promedio = juego / n

print(promedio)
```

$\therefore$  La simulación muestra que la ganancia promedio tiende a 0 conforme aumenta el número de repeticiones, por lo que el Juego 1 es un juego justo.

# Juego 2

```python
import numpy as np

cartas = np.arange(52)
print(cartas)
```

Cada 13 números vuelve a empezar el mismo valor
pero cambia el palo automáticamente

```python
(cartas % 13) + 1
```

Se calcula el valor esperado REAL matemático, es decir, se están revisando todas las combinaciones posibles.

La probabilidad de cada par es:
$$
p = (\frac{1}{52})(\frac{1}{51})
$$

```python
Ej = 0

valores = (cartas % 13) + 1
p = (1/52)*(1/51)
for ci in valores:
  for cj in valores:
    if ci != cj:
      if ci+cj >= 15:
        Ej += p
      else:
        Ej -= p

print(Ej)
```

Se va jugar un millón de veces.

```python
n = 10**6

suma = 0

for i in range(n):
  c1 = int(random.random()*52) + 1  # carta uno en [1,52] Cada ejecución te dará valores diferentes, todos entre 1 y 52. (pone el +1, para que el rango vaya de 1 a 52, y no de 0 a 52)
  v1 = (c1 % 13) + 1                # valor en [1,13] un número entero aleatorio entre 1 y 13
  mv = int(random.random()*51) + 1  # maximo valor en [1,51] que se sumará a c1
  c2 = ( (c1 +  mv) % 52 ) + 1      # carta dos entre [1,52] sin repetir c1
  v2 = (c2 % 13) + 1                # valor en [1,13]
  if v1 + v2 >= 15:
    suma += 1
  else:
    suma -= 1

print("La probabilidad estimada es: ", suma/n)
```

Es decir, en promedio pierdes ≈ 0.077 pesos por cada jugada.
Por ejemplo, al jugar 1000 veces pierdes 77 pesos.

# Conclusión

El Juego 1 es un juego justo ya que la ganancia esperada tiende a cero.

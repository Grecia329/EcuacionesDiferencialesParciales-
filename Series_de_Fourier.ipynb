Serie de Fourier a trozos

Dada la ecuacion a trozos

  f(x)=
  \begin{cases}
  0, & \text{si} \quad  -2\leq x<0\\
   2 -x, & \text{si} \quad 0 \leq x<2 \\
\end{cases}
  
La solucion de la serie de Fourier para f(x) esta dada por:
$$f(x)= \frac{1}{2} + \sum_{n=1}^{\infty} \left( \frac{2}{(n\pi)^2} (1 - (-1)^n) \cos \left( \frac{n\pi x}{2} \right) +  \frac{2}{n\pi} sen ( \frac{n\pi x}{2} \right).$$

Este problema modela la transferencia de calor en una barra ideal delgada, con $k=1$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Para empezar, graficamos la función definida a trozos.

```python
def f_x(x):
    return np.piecewise(x,[(-2 <= x) & (x<0), (0<=x) & (x<=2)],[0, lambda x: 2-x])
```

```python
x = np.linspace(-2,2,1000)
f = f_x(x)
plt.plot(x, f, label='$f(x)$', color='red', linewidth=2)

plt.xlabel('$x$')
plt.ylabel('$f(x)$')

# Título de la gráfica
plt.title("Gráfica de la Función a Trozos")

# Configuración de los ejes y leyenda
plt.xlim([-2.1,2.1])
plt.ylim([min(f) - 0.5, max(f) + 0.5])  # Ajuste dinámico del eje Y
plt.grid()
plt.legend()

# Mostrar la gráfica
plt.show()  # Asegurar que se llame plt.show()
```

Usamos una serie de Fourier para aproximar la función a trozos.

```python
def fourier_series(x, N=15):
    sum_cos = np.zeros_like(x)
    sum_sin = np.zeros_like(x)

    for n in range(1, N+1):
        sum_cos += (1-(-1)**n / n) * np.cos(n * np.pi * x / 2)
        sum_sin += (1 / n) * np.sin(n * np.pi * x / 2)
    return (1/2) + (2/np.pi**2) * sum_cos + (2/np.pi) * sum_sin

# Definir valores de x en [-2,2]
x = np.linspace(-2, 2, 1000)
```

```python
f=f_x(x)
f_fourier = fourier_series(x, N=15)  # Ajusta N para mayor precisión

# Graficar la función original
plt.plot(x, f, label='$f(x)', color='red', linewidth=2)

# Graficar la serie de Fourier
plt.plot(x, f_fourier, label='Serie de Fourier (N=100)', color='blue', linestyle='--')

# Etiquetas de los ejes y título
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Aproximación de la función con su Serie de Fourier")

# Configuración de los ejes y leyenda
plt.xlim([-2.1, 2.1])
plt.ylim([min(f) - 0.5, max(f) + 0.5])  # Ajuste dinámico del eje Y
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Mostrar la gráfica
plt.show()
```

Ahora, agregamos la parte temporal para visualizar la solución completa. Primero, para el momento inicial, $t=0$.

```python
def u_x_t(x,t,N=50):
    f = zeros( (len(x),len(t)) )
    for i in range(len(t)):
        for n in range(1,N):
            f[:,i] += (-1)**(n+1)*sin(n*pi*x)/(n)*exp(-n**2*pi**2*t[i])
    return (2.0/pi)*f
```

```python
x = linspace(0,1,1000)
t = linspace(0,10,1000)
U = u_x_t(x,t)
```
